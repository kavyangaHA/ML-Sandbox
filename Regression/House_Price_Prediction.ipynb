Importing Dependencies

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

Let's pretend we have a CSV file names "housing.csv"
data = pd.read_csv('housing.csv')
for this example,we will create a fake one


```python
data = {
    'square-footage':[1500,2500,1200,3000,1800],
    'num_bedrooms':[3,4,2,5,3],
    'num_bathrooms':[2,2,1,3,2],
    'neighborhood':['Suburb','Downtown','Suburb','Downtown','rural'],
    'price':[300000,550000,350000,600000,400000]
}
```

```python
df = pd.DataFrame(data)
```

```python
#preprocessing
#seperate features (x) and target (y)

x =df.drop('price',axis=1);
y = df['price']
```

```python
from seaborn import categorical
#Identify categorical features
categorical_feature = ["neighborhood"]
numerical_features = ["square-footage","num_bedrooms","num_bathrooms"]
```

```python
#create a preprocessor objet
#this will apply one - hat encoding to categorical features
#and leave numerical features as they are ('passthrouh)

preprocesssor  = ColumnTransformer(transformers=[('num','passthrough',numerical_features),('cat',OneHotEncoder(),categorical_feature)
                                                 ])
```

```python
#split data
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model = LinearRegression()
```

```python
#create a pipline that first preprocess the data,then fits the model
from sklearn.pipeline import Pipeline
model_pipeline = Pipeline(steps=[('preprocessor',preprocesssor),('regressor',model)])
```

```python
#train the model in the training data
model_pipeline.fit(X_train,y_train)
```

```python
#Evaluation
#Make predictions on the test set
y_pred = model_pipeline.predict(X_test)
```

```python
mae = mean_absolute_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)
```

```python
print(f"Model Evaluation:")
print(f"Mean Absolute Error (MAE):{mae:.2f}")
print(f"R-squared(R**2): {r2:.2f}")
```

```python
#let's predict the price of a new house
new_house = pd.DataFrame({'square-footage':[2000],'num_bedrooms':[3],'num_bathrooms':[2],'neighborhood' :'Suburb'})
```

```python
predicted_price = model_pipeline.predict(new_house)
```

```python
print(f"\nPredicted price for new house: ${predicted_price[0]:,.2f}")
```
